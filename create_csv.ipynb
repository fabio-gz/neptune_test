```python
import pandas as pd
```

```python
data_vertex = pd.read_excel('Matriz_de_adyacencia_data_team.xlsx', sheet_name=1, engine='openpyxl', header=3)
data_vertex.head()
```

```python
vertex = data_vertex.iloc[:,1:]
vertex.columns = ['~id', 'name:string']
```

```python
data_edges = pd.read_excel('Matriz_de_adyacencia_data_team.xlsx', sheet_name=0, engine='openpyxl')
data_edges.head()
```

```python
data_edges.drop('Unnamed: 1', axis=1, inplace=True)
```

```python
data_edges.drop([0], axis=0, inplace=True)
data_edges.head()
```

```python
data_edges = data_edges.iloc[:17,:].set_index('Unnamed: 0')
data_edges.head()
```

```python
df = data_edges.stack().reset_index()
df.head(15)
```

```python
edges = df[df[0] == 1]
edges['~id'] = range(1, len(edges) + 1)
edges.set_index('~id', inplace=True)
```

```python
edges = edges.iloc[:,:2]
edges.columns = ['~from', '~to']
```

```python
edges = edges.astype('int64')
```

```python
edges['~label'] = 'Relacion'
edges.head()
```

```python
edges.to_csv('edges.csv')
```
